```python
from tensorflow.keras.layers import Conv1D,BatchNormalization,LeakyReLU,MaxPool1D,\
GlobalAveragePooling1D,Dense,Dropout,AveragePooling1D
from tensorflow.keras.models import Sequential
from tensorflow.keras.backend import clear_session
def cnnmodel():
    clear_session()
    model=Sequential()
    model.add(Conv1D(filters=5,kernel_size=3,strides=1,input_shape=(1510,23)))#1
    model.add(BatchNormalization())
    model.add(LeakyReLU())
    model.add(MaxPool1D(pool_size=2,strides=2))#2
    model.add(Conv1D(filters=5,kernel_size=3,strides=1))#3
    model.add(LeakyReLU())
    model.add(MaxPool1D(pool_size=2,strides=2))#4
    model.add(Dropout(0.5))
    model.add(Conv1D(filters=5,kernel_size=3,strides=1))#5
    model.add(LeakyReLU())
    model.add(AveragePooling1D(pool_size=2,strides=2))#6
    model.add(Dropout(0.5))
    model.add(Conv1D(filters=5,kernel_size=3,strides=1))#7
    model.add(LeakyReLU())
    model.add(AveragePooling1D(pool_size=2,strides=2))#8
    model.add(Conv1D(filters=5,kernel_size=3,strides=1))#9
    model.add(LeakyReLU())
    model.add(GlobalAveragePooling1D())#10
    model.add(Dense(1,activation='sigmoid'))#11
    
    model.compile('adam',loss='binary_crossentropy',metrics=['accuracy'])
    return model

model=cnnmodel()
model.summary()
```

```python
from tensorflow.keras.optimizers import Adam

def cnnmodel_tuned():
    clear_session()
    model = Sequential()
    model.add(Conv1D(filters=5, kernel_size=3, strides=1, input_shape=(1510, 23)))
    model.add(BatchNormalization())
    model.add(LeakyReLU())
    model.add(MaxPool1D(pool_size=2, strides=2))
    model.add(Conv1D(filters=5, kernel_size=3, strides=1))
    model.add(LeakyReLU())
    model.add(MaxPool1D(pool_size=2, strides=2))
    model.add(Dropout(0.25))  # dropout rate
    model.add(Conv1D(filters=5, kernel_size=3, strides=1))
    model.add(LeakyReLU())
    model.add(AveragePooling1D(pool_size=2, strides=2))
    model.add(Dropout(0.25))  # dropout rate
    model.add(Conv1D(filters=5, kernel_size=3, strides=1))
    model.add(LeakyReLU())
    model.add(AveragePooling1D(pool_size=2, strides=2))
    model.add(Conv1D(filters=5, kernel_size=3, strides=1))
    model.add(LeakyReLU())
    model.add(GlobalAveragePooling1D())
    model.add(Dense(1, activation='sigmoid'))

    
    optimizer = Adam(learning_rate=0.0001, beta_1=0.6)

    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
    return model
```

```python
model=cnnmodel_tuned()
model.summary()
```

```python
from sklearn.model_selection import GroupKFold,LeaveOneGroupOut
from sklearn.preprocessing import StandardScaler
gkf=GroupKFold(n_splits=3)
```

```python
accuracy=[]
for train_index, val_index in gkf.split(data_array, label_array, groups=group_array):
    train_features,train_labels=data_array[train_index],label_array[train_index]
    val_features,val_labels=data_array[val_index],label_array[val_index]
    scaler=StandardScaler()
    train_features = scaler.fit_transform(train_features.reshape(-1, train_features.shape[-1])).reshape(train_features.shape)
    val_features = scaler.transform(val_features.reshape(-1, val_features.shape[-1])).reshape(val_features.shape)
    model=cnnmodel_tuned()
    model.fit(train_features,train_labels,epochs=10,batch_size=32,validation_data=(val_features,val_labels))
    accuracy.append(model.evaluate(val_features,val_labels)[1])
```

# Normalizing channel wise and conv to spectrogram

```python
from skimage.transform import resize
from scipy.signal import spectrogram
import os
import matplotlib.pyplot as plt
import numpy as np

def compute_spectrogram(data, fs):
    f, t, Sxx = spectrogram(data, fs=fs)
    return f, t, Sxx

num_samples = 20000  # Set the desired number of samples to 20,000

# Set the desired size for resizing
target_size_resized = (256, 256)

# Create a folder to save resized spectrogram images
output_folder_resized = '/home/rs/manashi.saharia/Priyadarshi/Spec_imgs_1/Resized_imgs_additional'
os.makedirs(output_folder_resized, exist_ok=True)

# Randomly sample from each label
selected_indices_label_0 = np.where(label_array == 0)[0][:num_samples // 2]
selected_indices_label_1 = np.where(label_array == 1)[0][:num_samples // 2]

# Combine the selected indices
selected_indices = np.concatenate([selected_indices_label_0, selected_indices_label_1])

# Shuffle the combined indices
np.random.shuffle(selected_indices)

for i, selected_index in enumerate(selected_indices):
    current_data = data_array[selected_index]
    current_label = label_array[selected_index]

    # Check if data has the expected shape (num_samples, sample_length, num_channels)
    if len(current_data.shape) == 3 and current_data.shape[0] > 0 and current_data.shape[1] > 0 and current_data.shape[2] > 0:
        # Normalize each channel independently
        for channel in range(current_data.shape[2]):
            current_channel_data = current_data[:, :, channel]
            current_data[:, :, channel] = (current_channel_data - np.mean(current_channel_data)) / np.std(current_channel_data)

        # Compute the spectrogram
        f, t, Sxx = compute_spectrogram(current_data, fs=1.0)
        Sxx_reshaped = 10 * np.log10(Sxx)

        # Resize the spectrogram image
        resized_image = resize(Sxx_reshaped, target_size_resized)

        # Save the resized spectrogram image with the corresponding label
        filename_resized = os.path.join(output_folder_resized, f'spectrogram_sample_{i + 1}_label_{current_label}.png')
        plt.imsave(filename_resized, np.expand_dims(resized_image, axis=-1), cmap='viridis')
    else:
        print(f"Skipping sample {i + 1} as it does not have the expected shape.")
```
